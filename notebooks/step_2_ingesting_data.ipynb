#### 3.1 : Creating new columns from Features column

```python
def create_column_dict(substring,dicts):
    """function to create dictionary from features column based on substring search
        Args:
        substring (str): string to be searched in each row of features column
        dicts (dict)   : dictionary to store string values
        Returns
        dicts (dict)   : dictionary with row number as key and string as value        
    """    
    for i in range(len(df)):
        for j in range(1,15):
            result = re.search(substring, features[i][j])
            if result!=None:
                dicts[i]=features[i][j]
                
    return dicts
```

```python
warranty=create_column_dict('Warranty',{})
```

```python
processor=create_column_dict('Processor',{})
display=create_column_dict('Display',{})
ram=create_column_dict('RAM',{})
os=create_column_dict('Operating System',{})
office=create_column_dict('Office',{})
```

```python
storage=create_column_dict('HDD',{})
create_column_dict('SSD',storage)
```

```python
def create_column(name,value):
    """function to create columns in dataframe df
        Args:
        name (str)     : Name of column
        value (dict)   : values to be stored in column              
    """ 
    df[name]=str()
    for i in value.keys():
        if value[i]!=None:
            df[name][i]=value[i]  
```

```python
column_name=['Processor','Storage','RAM','OS','Display','Office','Warranty']#name of column
column_value=[processor,storage,ram,os,display,office,warranty]#dict value of column
```

```python
#creating columns in dataframe df
for i in range(len(column_name)):
    create_column(column_name[i],column_value[i])    
```
